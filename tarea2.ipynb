```python
# Importar librerías
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. Analiza la base de datos `diabetes.csv`

```python
df = pd.read_csv('diabetes-1.csv')
df.head(10)
```

### Por lógica, sustituir los valores de cero por "none"

```python
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
df.head(10)
```

## 2. Grafica la densidad de probabilidad de las etiquetas `diabético = 1` y `No diabético = 0`

```python
db_0 = df[df.Outcome == 0] #Cuento las etiquetas con valor cero
db_1 = df[df.Outcome == 1] #Cuento las etiquetas con valor uno

for c in columns:
    sns.distplot(db_0[c],kde=True,  hist = False, label='Etiqueta = 0')
    plt.legend()
    sns.distplot(db_1[c], kde=True, hist = False, label='Etiqueta = 1')
    plt.legend()
    plt.show()
```

# 3. División de los datos

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
```

```python
X = df[columns]
Y = df['Outcome']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, random_state=1)
```

# 4. Creación y entrenamiento del modelo

```python
modelo = GaussianNB()
yPred = modelo.fit(X_train, Y_train).predict(X_test)
print('Exactitud:', accuracy_score(Y_test,yPred))
```

# 5. Grafique la curva ROC, el área bajo al curva y la matriz de confusión del modelo.

### Curva ROC y Área bajo la curva

```python
# Importar librerías
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
# Calcular el área bajo la curva (AUC)
r_probs = [0 for _ in range(len(Y_test))]
dt_probs = modelo.predict_proba(X_test)
dt_probs = dt_probs[:,1]

r_auc = roc_auc_score(Y_test, r_probs)
dt_auc = roc_auc_score(Y_test, dt_probs)

print('Área bajo la curva (AUC):', dt_auc)

# Graficar la curva ROC y el área bajo la curva
r_fpr, r_tpr, _ = roc_curve(Y_test, r_probs)
dt_fpr, dt_tpr, _ =roc_curve(Y_test,dt_probs)

plt.plot(r_fpr, r_tpr, linestyle='--', label='Prediccion Aleatoria (AUROC = %0.3f)' % r_auc)
plt.plot(dt_fpr, dt_tpr, marker ='.', label ='Decision Tree ( AUROC = %0.3f)' % dt_auc)

# Agregar el área bajo la curva (AUC) al gráfico
plt.fill_between(dt_fpr, dt_tpr, alpha=0.3)

plt.title('Curva ROC')
plt.xlabel('Tasa de falsos positivos')
plt.ylabel('Tasa de verdaderos positivos')
plt.legend()
plt.show()
```

### Matriz de confusión

```python
class_names = ['Outcome']

sns.heatmap(confusion_matrix(Y_test,yPred), annot=True, fmt='3.0f')
plt.xlabel('Predicted Labels')
plt.ylabel('True labels')
plt.title('Confusion matrix', y=1.05, size=15)
```

# 6. Grafique la curva de aprendizaje del modelo y ajuste los valores para obtener un entrenamiento óptimo.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.model_selection import learning_curve

# Utilizar la función de trazado de la curva de aprendizaje
train_sizes, train_scores, test_scores = learning_curve(modelo, X_train, Y_train)
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)
plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                    train_scores_mean + train_scores_std, alpha=0.1,
                    color="r")
plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                    test_scores_mean + test_scores_std, alpha=0.1, 
                    color="g")
plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
            label="Score de entrenamiento")
plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
            label="Score de validación")
plt.title('Curva de aprendizaje')
plt.xlabel("Ejemplos de entrenamiento")
plt.ylabel("Score")
plt.legend()
plt.show()
```
